# Advanced part of WWR distribution
In the last turtorial, the windows are propotional distributed on walls.However, in the real world buildings, windows often have a fixed height 1.5m. Therefore, WWR affects the width of windows. Moreover, there could be several windows on a wall, ideally, there are uniformly distributed.

### Step 2 : Import a baseline model

We import a baseline model that will be used for simulations. You can download from [icepy github page](https://github.com/chenglongfu11/icepy/tree/master/buildings). 

```python
folder = 'D:\\ide_mine\\changing\\ut2pro\\'
base_model = 'ut2proportional'
```

The wall width list and the building height are provided along with the idm model.

```python
wall_width_list = [4, 7.5, 10.66, 10.0, 17.97, 21.64, 6.528, 7.71, 6.528, 4, 6.516, 1.6, 30, 3.992, 8.43, 8.295,
                  14.325, 10.88, 8.52]
building_ht = 22.5 
```

You can set desired the floor height, the number of floors.

```python
floor_ht = 3                 
num_floors = 5
```

Let's import the idm model

```python
from icepy.modeleditor import IDM
idm = IDM(folder, base_model)
```

### Step 3: Add windows according to WWR ratio

```python
wwrs = [0.1,0.2,0.25, 0.3, 0.35,0.4,0.45, 0.47]
```

Options: 0: proportional distribution   1: fixed height 1.5m  2: multiple standard windows

```python
new_files = []
for wwr in wwrs:
    idm1 = IDM(folder, base_model)
    idm1.openfile(base_model)
    new_file = idm1.applywwr_new(wwr, wall_width_list, floor_ht, base_model, option=0)
    new_files.append(new_file)
    idm1.close()
```

Check if all the new files with new wwrs are generated correctly

```python
new_files
```

```python
import os.path
for elem in new_files:
    print(os.path.exists(elem))
```

#### Step 3 extra: Solve error and generate models with correspoding wwrs manually
Due to the unreliablity of IDA ICE, there could be some errors or mistakes causing the model missing. Please make sure you keep IDA ICE on foreground, monitor the process and check the error message.  

```python
remain_wwr = [0.1,0.35]
```

```python
for wwr in remain_wwr:
    idm1 = IDM(folder, base_model)
    idm1.openfile(base_model)
    new_file = idm1.applywwr_new(wwr, wall_width_list, floor_ht, base_model, option=0)
    idm1.close()
```

### Step 4: Clone zones

In the idm model, one zone with 2.6 height for the ground floor is included. We assume there is one zone for each floor. So we clone zones. Note that the collection of zones should not overflow this building.

```python
new_file_paths =[]
for ele in new_files:
    idm2 = IDM(folder, ele)
#     print(idm2.folder)
#     print(idm2.fname)
    idm2.openfile_path(ele)
    new_file = idm2.zoneclone_save(floor_ht, num_floors)
    new_file_paths.append(new_file)
    idm2.close()
```

Check if all the new files with new wwrs are generated correctly

```python
new_file_paths
```

```python
import os.path
for elem in new_file_paths:
    print(os.path.exists(elem))
```

#### Step 4 extra: Solve error and generate models with correspoding floors manuallyÂ¶
Due to the unreliablity of IDA ICE, there could be some errors or mistakes causing the model missing. Please make sure you keep IDA ICE on foreground, monitor the process and check the error message.

```python
new_files = ['D:\\ide_mine\\changing\\ut2pro\\ut2proportional_wwr0.1.idm']
```

```python
for ele in new_files:
    idm2 = IDM(folder, ele)
    idm2.openfile_path(ele)
    new_file = idm2.zoneclone_save(floor_ht, num_floors)
    new_file_paths.append(new_file)
    idm2.close()
```

### Step 5 : Simulate models sequentally

```python
computational, html_dicts = idm.sequential_simulation_wwr(new_file_paths, wwrs)
computational
```

## Part 2: Apply windows with multiple standard windows

The only difference with the part 1 is step 3 

### Step 3: Add windows according to WWR ratio

```python
wwrs = [0.3, 0.4, 0.45]
```

Options: 0: proportional distribution 1: fixed height 1.5m 2: multiple standard windows

```python
new_files = []
for wwr in wwrs:
    idm1 = IDM(folder, base_model)
    idm1.openfile(base_model)
    new_file = idm1.applywwr_new(wwr, wall_width_list, floor_ht, base_model, option=2)
    new_files.append(new_file)
    idm1.save_close()
```

## Part 3: Results retrive and database
Simulation results can be exported into html files and read. You are free to use MySQL database to store these results.


The computational results are time records of simulations. 

```python
computational
```

The html dictionarties are wwrs and corresponding html report paths. We use them to get simulation results

```python
html_dicts
```

```python
from icepy.readhtml import Readhtml
readHTML = Readhtml()
```

We can get simulation results from the html paths. The valueable results are energy consumption of grand total (kwh) and energy consumtion per square of grand total(kwh/m2). The generated results are in dataframe format. 

```python
# Retrieve the dataframe of deliverd energy
df_tot = readHTML.multi_df_html(html_dicts)
```

```python
df_tot[0]
```

Now we insert simulation results into the mysql databse.

```python
import icepy.database as database
```

```python
db = database.DB()
```

You can define 'note' by yourself to recognize data. There are three tables designed in mysql database, baseline, simulation_info, simu_results.

```python
note = 'proportional'
db.insert_computational_df_html(base_model, building_ht, folder, floor_ht, 
                                num_floors, new_files[0], note, computational, df_tot, html_dicts)
```

```python
for ht in html_dicts:
    ht['html'] = ht['html'][0:40]
```
